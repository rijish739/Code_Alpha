```python
import speech_recognition as sr

# Initialize recognizer
recognizer = sr.Recognizer()

# Load audio file
audio_file = "WhatsApp Audio 2025-03-02 at 12.11.32_1fc9d74c.waptt.wav"  # Replace with your audio file

with sr.AudioFile(audio_file) as source:
    audio_data = recognizer.record(source)  # Record audio
    text = recognizer.recognize_google(audio_data)  # Convert speech to text

print("Recognized Text:", text)
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load an example RAVDESS audio file
file_path = "WhatsApp Audio(2) 2025-03-02 at 12.22.34_a9ad1886.wav"
y, sr = librosa.load(file_path, sr=16000)

# Display waveform
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr)
plt.title("RAVDESS Audio Waveform")
plt.show()
```
