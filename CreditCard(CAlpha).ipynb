```python
x = df.drop(columns   = ['Y'], axis = 1)
y = df['Y']
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
y_train
```

```python
print(y_train.dtype)  # Should be int, not float or object
print(type(y_train))  # Should be a pandas Series or NumPy array

```

```python
print(y_test.dtype)  # Should be int, not float or object
print(type(y_test))  # Should be a pandas Series or NumPy array
```

```python
import numpy as np
y_train = np.array(y_train, dtype=int)  # Ensures integer labels
```

```python
import numpy as np
y_test = np.array(y_test, dtype=int)  # Ensures integer labels
```

```python
scaler = StandardScaler()
scaler.fit(x_train)
```

```python
x_train = scaler.fit_transform(x_train)

x_test = scaler.transform(x_test)
print(x_train)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(y_train, x_train_prediction)
print("Accuracy score: ", training_data_accuracy*100 )
print(x_train_prediction)
```

```python
conf_matrix = confusion_matrix(y_train, x_train_prediction)
classification = classification_report(y_train, x_train_prediction)
print("confusion_matrix: \n", conf_matrix)
print("Accuracy score of LR model is: ",training_data_accuracy)
print("classification_report: \n", classification)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(y_test, x_test_prediction)
print("Accuracy score: ", train_data_accuracy*100 )
print(x_test_prediction)
```

```python
conf_matrix = confusion_matrix(y_test, x_test_prediction)
classification = classification_report(y_test, x_test_prediction)
print("confusion_matrix: \n", conf_matrix)
print("Accuracy score of LR model is: ",test_data_accuracy)
print("classification_report: \n", classification)
```
